# 0.3 Mixed Types 

```python
#Checking for mixed types
for col in df.columns.tolist():
    weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis=1)
    if len (df[weird])>0:
        print(col)
```

#### Female to Male ratio is not essential column, types mixing does not constitute a problem

```python
df.dtypes
```

#### The column International Students is expressed in %, Female to Male Ratio is expressed in xx:xx. These are strings, other numerical columns are in numeric format as they should.
